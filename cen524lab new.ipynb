```python
import math

def geometricprogression(a,r,infinite):
    if infinite:
         return ( (a) / (1 - r))

    n = int(input("Enter number of terms:"))
    if r == 1:
        return a * n
    elif r < 1:
        return ( ((a) * (1 - math.pow(r,n))) / (1-r))
    else:
        return ( ((a) * (math.pow(r,n)) - 1 ) / (r - 1))

print(geometricprogression(8,0.2,True))

print('GP works.')
```
